```python
#Tristimulus values 

def tristimulus_vals(P, ciex, ciey, ciez):
    X = P*ciex
    Y = P*ciey
    Z = P*ciez
    wav = np.linspace(380,750, num=75)
    X_area = np.trapz(X, wav)
    Y_area = np.trapz(Y, wav)
    Z_area = np.trapz(Z, wav)
    return X_area, Y_area, Z_area
    
```

```python
#CIE xyz 

def cie_xyz(X, Y,Z):
    I = X + Y + Z
    x = X/I
    y = Y/I
    z = 1 - x- y
    return x,y
```
